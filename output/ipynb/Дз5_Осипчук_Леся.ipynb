```python
import matplotlib.pyplot as plt
from sympy import diff, symbols
import numpy as np

fig, ax = plt.subplots()

# Перемістимо лівий і нижній стовпчики до x = 0 і y = 0 відповідно.
ax.spines[["left", "bottom"]].set_position(("data", 0))

# Сховати верхню та праву лінію
ax.spines[["top", "right"]].set_visible(False)

# Намалюємо стрілки (як чорні трикутники: ">k"/"^k") на кінцях осей.
# Також вимкнемо відсікання (clip_on=False) стрілок.
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

# Додамо проміжні лінії
ax.grid(True, linestyle='-.')

x, y = symbols('x y')
# Задаємо функцію
fx = 1/3*x**3+1/2*x**2-2*x

# Обраховуємо похідну функції
dx = diff((fx))
print (dx)
k = []
n = []


for i in range(1000):
    j = (i-300)*0.01
    k.append(j)
    n.append(dx.subs(x,j))

# Значення x в заданих точках
x_values = [-1/2, 1]

# Отримання значень похідної функції при заданих точках x
for x_val in x_values:
    # Підставлення значення x у вираз для похідної функції dx
    derivative_value = dx.subs(x, x_val)
    # Друк результатів
    print(f"f'({x_val}) = {derivative_value}")

ax.plot(k,n)


# Запускаємо малювання графіку
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import seaborn as sns

fig, ax = plt.subplots()

# Перемістимо лівий і нижній стовпчики до x = 0 і y = 0 відповідно.
ax.spines[["left", "bottom"]].set_position(("data", 0))

# Сховати верхню та праву лінію
ax.spines[["top", "right"]].set_visible(False)

# Намалюємо стрілки (як чорні трикутники: ">k"/"^k") на кінцях осей.
# Також вимкнемо відсікання (clip_on=False) стрілок.
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

# Додамо проміжні лінії
ax.grid(True, linestyle='-.')

# x, y = sp.symbols ('x y')
x = sp.Symbol('x')
# Задаємо функцію
f = -3*x**2+30*x

sns.scatterplot(x=[5], y=[75], ax=ax, color = 'red')
# Обраховуємо похідну функції
df = f.diff()
print (df)

k = []
n = []
l = []

for i in range(1000):
    j = (i-300)*0.01
    k.append(j)
    n.append(df.subs(x,j))
    l.append(f.subs(x,j))


ax.plot(k,n)
ax.plot(k,l)



# Запускаємо малювання графіку
plt.show()
```
