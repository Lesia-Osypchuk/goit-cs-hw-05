```python
def stock_price_at_time(t):
    return np.random.gamma(0.3, 1.1, t).sum()

def simulate_n_times(n, t):
    np.random.seed(42)
    prices = []
    for _ in range(n):
        prices.append(stock_price_at_time(t))
    return prices

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import normaltest

n_simulations = 100
for t in range(1, 61, 2):
    prices = simulate_n_times(n_simulations, t)
    average_price = np.mean(prices)
    print(f"Середня ціна для {n_simulations} симуляцій з t = {t}: {average_price:.2f}")



    plt.figure(figsize=(8, 6))
    plt.hist(prices, bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Гістограма ціни для {n_simulations} симуляцій з t = {t}')
    plt.xlabel('Ціна')
    plt.ylabel('Частота')
    plt.grid(True)
    plt.show()


    stat, p_value = normaltest(prices)
    if p_value > 0.05:
        print(f"p-значення тесту normaltest для t = {t}: {p_value:.4f}. Розподіл нормальний.")
    else:
        print(f"p-значення тесту normaltest для t = {t}: {p_value:.4f}. Розподіл не нормальний.")
```

Можна зробити висновки, що при збільшенні t (часу):
- середні ціни відображають тенденцію до зростання;
- розподіл цін стає більш асиметричним;
- для більшості значень t розподіл не є нормальним.

Це вказує на те, що розподіл цін акцій не може бути точно відтворений гамма-розподілом при довших періодах торгівлі.


```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import normaltest

# Імпорт даних з файлу
url = "https://docs.google.com/spreadsheets/d/18WCpPS96Tb3cB0FCsIA92PEhcmBkp08sjYhS9DsQfJE/export?format=csv"

df = pd.read_csv(url)

# Середнє значення, дисперсія та стандартне відхилення для кожної колонки

summary_stats = df.describe().loc[['mean', 'std']].T
summary_stats.columns = ['mean', 'std_deviation']

# Розрахунок дисперсії для кожної колонки
variance = df.var()

# Додати дисперсію до таблиці зі статистикою
summary_stats['variance'] = variance

# Побудова гістограм для кожної колонки
for column in df.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=column, kde=True, bins=20)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

# Перевірка на нормальність розподілу
normality_test_results = {}
for column in df.columns:
    stat, p = normaltest(df[column])
    normality_test_results[column] = {'Test Statistic': stat, 'p-value': p}


# Кореляція з Product_Sold
correlation_with_product_sold = df.corr()['Product_Sold']

# Друк результатів
print("Дані для кожного показника:середнє значення, стандартне відхилення, дисперсія")
print(summary_stats)
print("\nNormality Test Results:")
for column, result in normality_test_results.items():
    print(f"{column}: Test Statistic = {result['Test Statistic']}, p-value = {result['p-value']}")
    if result['p-value'] > 0.05:
        print(f"{result['p-value']:.4f} Розподіл нормальний.")
    else:
        print(f" {result['p-value']:.4f} Розподіл не нормальний.")
print("\nКореляція з Product_Sold:")
print(correlation_with_product_sold.values)

```
